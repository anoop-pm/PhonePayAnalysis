```python
import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Payment_Mode':[], 'Total_Transactions_count':[], 'Total_Amount':[]}
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Payment_Mode'].append(Name)
              clm['Total_Transactions_count'].append(count)
              clm['Total_Amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe

print(D)
Agg_Trans=pd.DataFrame(clm)
Agg_Trans
```

```python
Agg_Trans.to_csv("Data_Aggregated_Transaction_Table.csv", index=False)
```

```python


path3="/content/pulse/data/map/user/hover/country/india/state/"

Map_state_list=os.listdir(path3)
Map_state_list

```

```python

# Data_Map_IndiaStates_TU
clm4={'state':[], 'Registered_Users':[]}
for i in Map_state_list:
    p_i=path3+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)

            users_by_device = D['data']
      
            registered_users = sum([d['registeredUsers'] for d in users_by_device['hoverData'].values()])

            clm4['Registered_Users'].append(registered_users)
         
            clm4['state'].append(i)
    print(i)

#Succesfully created a dataframe
Map_User=pd.DataFrame(clm4)

```

```python

duplicate_sum = Map_User.groupby('state')['Registered_Users'].sum().reset_index()

# display the resulting dataframe
duplicate_sum
```

```python
duplicate_sum.to_csv("Data_Map_IndiaStates_TU.csv", index=False)
```

```python
# Place_Name,Total_Transactions_count,Total_Amount,Quarter,Year,State


path4="/content/pulse/data/map/transaction/hover/country/india/state/"

Map_state_list_tr=os.listdir(path4)
Map_state_list_tr
```

```python
clm5={'State':[], 'Year':[],'Quater':[],'Place_Name':[], 'Total_Transactions_count':[],'Total_Amount':[]}

for i in Map_state_list_tr:
    p_i=path4+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            user_datas=D['data']['hoverDataList']
            for device in user_datas :
                Place_Name = device['name']
                new_metric=device['metric']
                for device2 in new_metric :
                
                  Total_Transactions_count = device2['count']
                  Total_Amount = device2['amount']
                  clm5['Place_Name'].append(Place_Name)
                  clm5['Total_Transactions_count'].append(Total_Transactions_count)
                  clm5['Total_Amount'].append(Total_Amount)
                  clm5['State'].append(i)
                  clm5['Year'].append(j)
                  clm5['Quater'].append(int(k.strip('.json')))


            
            
#Data_Map_Transaction_Table
Map_Transaction=pd.DataFrame(clm5)
Map_Transaction
```

```python
# Data_Map_User_Table




path5="/content/pulse/data/map/user/hover/country/india/state/"

Map_state_list_users=os.listdir(path5)
Map_state_list_users
```

```python
clm6={'State':[], 'Year':[],'Quater':[],'Place_Name':[], 'Registered_Users_Count':[],'App_Openings':[]}

for i in Map_state_list_users:
    p_i=path5+i+"/"
    Agg_yr=os.listdir(p_i)    
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)        
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            user_datas=D['data']['hoverData']
            dictiona=dict(user_datas)
            Place_Name=None
            for key in dictiona.keys():
              Place_Name=key
              
              reg_users=user_datas[Place_Name]['registeredUsers']
              App_Openings=user_datas[Place_Name]['appOpens']

   
              clm6['Place_Name'].append(Place_Name)
              clm6['Registered_Users_Count'].append(reg_users)
              clm6['App_Openings'].append(App_Openings)
              clm6['State'].append(i)
              clm6['Year'].append(j)
              clm6['Quater'].append(int(k.strip('.json')))


```

```python
# Data_Map_User_Table
User_map_data=pd.DataFrame(clm6)
User_map_data
```

```python

User_map_data.to_csv("Data_Map_User_Table.csv", index=False)
```
